```python
##### helper methods 
def reset_time(G):
    reset = 0
    
    n = G.order()
    
    for i in range(n):
        reset += stationary_distribution(G,i) * mixing_time_vertex(G,i)
    
    return reset

def best_mixing_time(G):
    n = G.order()
    min_mix = n^2
    for i in G.vertices():
        mix = mixing_time_vertex(G,i)
        if mix < min_mix:
            min_mix = mix
    return min_mix

def second_eigenvalue(G):
    trans_mat = transition_matrix(G)
    eigenvals = trans_mat.eigenvalues()
    eigenvals.sort()
    return eigenvals[-2]

def transition_matrix(G):
    n = G.order()
    adj_matrix = G.adjacency_matrix()
    trans_matrix = zero_matrix(QQ, n,n)
    for idx in range(n):
        row = adj_matrix.row(idx)
        degree = sum(row)
        trans_row = [ x / degree for x in row]
        trans_matrix.set_row(idx, trans_row)
    
    return trans_matrix


def from_mix_map(n,d,idx):
    tree_list = mix_map[(n,d)]
    #print(tree_list)
    return tree_list[idx]
    
### Minimum Reset Time
def min_reset(n, d, show=True):
    tree_list = from_mix_map(n,d,0)
    reset = reset_time(Graph(tree_list[0]))
    if show:
        print('MIN RESET\tn=', n, 'd=', d, 'reset=', reset)
        for t in tree_list:
            #print(t)
            graph = Graph(t)
            graph.show()
    return reset

### Maximum Reset Time
def max_reset(n, d, show=True):
    tree_list = from_mix_map(n,d,1)
    reset = reset_time(Graph(tree_list[0]))
    if show:
        print('MAX RESET\tn=', n, 'd=', d, 'reset=', reset)
        for t in tree_list:
            #print(t)
            graph = Graph(t)
            graph.show()
    return reset

### Minimum Mixing Time
def min_mixing(n, d, show=True):
    tree_list = from_mix_map(n,d,2)
    mix = mixing_time(Graph(tree_list[0]))
    if show:
        print('MIN MIXING\tn=', n, 'd=', d, 'mixing=', mix)
        for t in tree_list:
            #print(t)
            graph = Graph(t)
            graph.show()
    return mix
        
### Maximum Mixing Time
def max_mixing(n, d, show=True):
    tree_list = from_mix_map(n,d,3)
    mix = mixing_time(Graph(tree_list[0]))
    if show:
        print('MAX MIXING\tn=', n, 'd=', d, 'mixing=', mix)
        for t in tree_list:
            #print(t)
            graph = Graph(t)
            graph.show()
    return mix


def from_mix_map2(n,d,idx):
    tree_list = mix_map2[(n,d)]
    #print(tree_list)
    return tree_list[idx]

### Minimum Best Mixing Time
def min_bestmix(n, d, show=True):
    tree_list = from_mix_map2(n,d,0)
    bestmix = best_mixing_time(Graph(tree_list[0]))
    if show:
        print('MIN BESTMIX\tn=', n, 'd=', d, 'bestmix=', bestmix)
        for t in tree_list:
            #print(t)
            graph = Graph(t)
            graph.show()
    return bestmix

### Maximum Best Mixing Time
def max_bestmix(n, d, show=True):
    tree_list = from_mix_map2(n,d,1)
    bestmix = best_mixing_time(Graph(tree_list[0]))
    if show:
        print('MAX BESTMIX\tn=', n, 'd=', d, 'bestmix=', bestmix)
        for t in tree_list:
            #print(t)
            graph = Graph(t)
            graph.show()
    return bestmix

### Minimum Mixing Time
def min_eigenvalue(n, d, show=True):
    tree_list = from_mix_map2(n,d,2)
    eigen = second_eigenvalue(Graph(tree_list[0]))
    if show:
        print('MIN EIGENVALUE\tn=', n, 'd=', d, 'eigenvalue=', eigen)
        for t in tree_list:
            #print(t)
            graph = Graph(t)
            graph.show()
    return eigen
        
### Maximum Mixing Time
def max_eigenvalue(n, d, show=True):
    tree_list = from_mix_map2(n,d,3)
    eigen = second_eigenvalue(Graph(tree_list[0]))
    if show:
        print('MAX EIGENVALUE\tn=', n, 'd=', d, 'eigenvalue=', eigen)
        for t in tree_list:
            #print(t)
            graph = Graph(t)
            graph.show()
    return eigen
```
